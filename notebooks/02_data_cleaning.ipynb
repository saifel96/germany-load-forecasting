### Cleaning Entsoe data

```python
import pandas as pd
```

```python
load = pd.read_csv("/Users/saif/Desktop/ML_Load_Forecasting/data/Bronze/Entsoe_2020_2023.csv", index_col=0 , parse_dates= True)

load.index = pd.to_datetime(load.index, utc = True)
load.index = load.index.tz_localize(None)
load.columns = ["Load_MW"]
load = load.resample("1h").first()
load = load.iloc[1:]
```

```python
load.head()
```

```python
load.describe()
```

```python
load.info()
```

```python
load.corr()
```

```python
full_range = pd.date_range(
    start="2020-01-01 00:00:00",
    end="2023-12-31 23:00:00", 
    freq= "h"
)
```

```python
missing = full_range.difference(load.index)
```

```python
print(missing)
print(len(missing))
```

```python
load = load.reindex(full_range)
load.tail()
```

```python
load = load.interpolate()
load.tail()
```

```python
print(load.isna().sum())
```
