### Data Collecting From Entso-e and Openweatherdata -APIs

```python
import pandas as pd 
import os
from dotenv import load_dotenv
from entsoe import EntsoePandasClient
import requests
```

### Import Entso-e Data

```python
load_dotenv()
API_KEY = os.getenv('ENTSOE_API_KEY')

client = EntsoePandasClient(api_key=API_KEY)

start = pd.Timestamp('20200101', tz= "Europe/Berlin")
end = pd.Timestamp('20231231', tz= "Europe/Berlin")
country_code = 'DE_LU'
ts = client.query_load(country_code=country_code, start=start, end=end)

ts.to_csv("/Users/saif/Desktop/ML_Load_Forecasting/data/Bronze/Entsoe_2020_2023.csv")
```

### Import Weather data

```python
url = "https://archive-api.open-meteo.com/v1/archive"

params = {
    "latitude": 52.52,
    "longitude": 13.41,   # Berlin (you can change later)
    "start_date": "2020-01-01",
    "end_date": "2023-12-31",
    "hourly": [
        "temperature_2m",
        "wind_speed_10m",
        "shortwave_radiation"
    ],
    "timezone": "Europe/Berlin"
}

response = requests.get(url, params=params)
data = response.json()

df = pd.DataFrame(data["hourly"])
df["time"] = pd.to_datetime(df["time"])
df.set_index("time", inplace=True)

df.to_csv("/Users/saif/Desktop/ML_Load_Forecasting/data/Bronze/weather_2020_2023.csv")

print(df.head())
```
